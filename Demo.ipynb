```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import SearchEngine as SE
search_eng = SE.SearchEngine() # initialize a search engine
```

## Search Title

```python
boosting_factor = float(input("Enter to_read_boosting_factor ? :\n"))
```

```python
res = search_eng.query(user_id=904, text="i love",to_read_boosting_factor = boosting_factor, K=10_000, save_res=False) # we set K = 1e4 here just for demo purpose, 
# so we can see the tail of all books. This K is set to 20 by default, retrieving top 20 recommendations.
```

```python
res.head(20) # top recommended books
```

```python
res.tail() # least recommended books
```
